```python
contrast = pd.read_csv('/content/drive/MyDrive/AIEngineer/contrast/tmt/data center/for_nine.csv')
contrast.head(2)
```

```python
import pandas as pd
import ast

# Ensure 'name1' and 'name2' are strings for matching
contrast[['name1', 'name2']] = contrast[['name1', 'name2']].astype(str)

# Create a mapping dictionary from 'name1' and 'name2' to 'GPID1' and 'GPID2'
gpid_mapping = pd.concat([
    contrast[['name1', 'GPID1']].rename(columns={'name1': 'Drug', 'GPID1': 'GPID'}),
    contrast[['name2', 'GPID2']].rename(columns={'name2': 'Drug', 'GPID2': 'GPID'})
])

# Drop duplicates in case some drugs appear multiple times with the same GPID
gpid_mapping = gpid_mapping.drop_duplicates()

# Function to get GPID List from Drug List
def get_gpid_list(drug_list):
    drugs = [drug.strip() for drug in drug_list.split(',')]  # Split and clean drug names
    gpid_list = gpid_mapping[gpid_mapping['Drug'].isin(drugs)]['GPID'].dropna().tolist()
    return ', '.join(map(str, gpid_list)) if gpid_list else None

# Apply function to create new column 'GPID List'
test['GPID List'] = test['Drug List'].apply(get_gpid_list)
test
```

```python
# Save the updated dataframe
test.to_csv('/content/drive/MyDrive/AIEngineer/contrast/tmt/data center/test_with_gpid.csv', index=False)
```
