```python
# Load and split the text using tokenizer-aware chunking
with open('/content/drive/MyDrive/AIEngineer/agenda/company_1_better.txt', 'r') as file:
    text = file.read()
```

## **extract_keywords**

```python
from pythainlp.corpus import thai_stopwords
from pythainlp.tokenize import word_tokenize
from pythainlp.summarize.keybert import KeyBERT
from transformers import AutoTokenizer

# Function to remove stopwords from the text
def remove_stopwords(text, stop_words):
    # Tokenize the text
    words = word_tokenize(text, engine='deepcut')
    # Filter out stopwords
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

# Function to chunk the text into smaller pieces with a token limit
def chunk_text(text, tokenizer, max_tokens=512):
    # Tokenize the text
    tokens = tokenizer.tokenize(text)
    # Chunk tokens into smaller pieces
    return [tokenizer.convert_tokens_to_string(tokens[i:i + max_tokens]) for i in range(0, len(tokens), max_tokens)]

# Load Thai stopwords and add additional stopwords
stop_words = list(thai_stopwords())
additional_stop_words = ["นะคะ", "ครับ", "ก็", "โอเค", "ดีมาก", "ฮัลโหล", "กัน", "ที่", "คัน", "หรอก", "ไม่ต้อง", "ไม่", "น่าจะ", "คุ้นๆ", "แป๊บ", "นึง", "เดี๋ยว", "\n"]
stop_words.extend(additional_stop_words)
```

```python
# Remove stopwords
clean_text = remove_stopwords(text, stop_words)
clean_text
```

```python
# Initialize tokenizer (use the appropriate tokenizer for your model)
tokenizer = AutoTokenizer.from_pretrained("airesearch/wangchanberta-base-att-spm-uncased")
# Chunk the cleaned text
chunks = chunk_text(clean_text, tokenizer, max_tokens=500)
chunks
```

```python
# Initialize KeyBERT
kb = KeyBERT()

# Extract keywords from each chunk
all_keywords = []
for i, chunk in enumerate(chunks):
    keywords = kb.extract_keywords(chunk, max_keywords=10, return_similarity=True)
    print(f"Keywords for chunk {i + 1}: {keywords}")  # Print keywords for each chunk
    all_keywords.append(keywords)

combined_keywords = [keyword for sublist in all_keywords for keyword in sublist]
```
