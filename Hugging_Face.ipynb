```python

from transformers import BertForSequenceClassification, Trainer, TrainingArguments, AutoTokenizer
from datasets import load_dataset

# โหลดข้อมูลตัวอย่าง
dataset = load_dataset('glue', 'mrpc')

# โหลดโมเดลและ Tokenizer
model = BertForSequenceClassification.from_pretrained('bert-base-uncased')
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

# เตรียมข้อมูลสำหรับการฝึก
def tokenize_function(examples):
    return tokenizer(examples['sentence1'], examples['sentence2'], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)

# ตั้งค่า arguments สำหรับการฝึก
training_args = TrainingArguments(
    output_dir='./results',
    evaluation_strategy="epoch",
    learning_rate=2e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=3,
    weight_decay=0.01,
)

# สร้าง Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['validation']
)

# ฝึกโมเดล
trainer.train()
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import load_dataset
import numpy as np
import os

# โหลดโมเดลและ tokenizer
model_name = "airesearch/wangchanberta-base-att-spm-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3)

# ฟังก์ชันสำหรับเตรียมข้อมูล
def prepare_data():
    dataset = load_dataset("sepidmnorozy/Thai_sentiment")
    train_dataset = dataset['train']

    print("Columns in the dataset:", train_dataset.column_names)
    print("Sample data:", train_dataset[0])

    def convert_label(example):
        label_map = {"positive": 2, "neutral": 1, "negative": 0}
        if isinstance(example["label"], str):
            example["label"] = label_map.get(example["label"].lower(), 1)
        return example

    train_dataset = train_dataset.map(convert_label)

    def tokenize_and_align_labels(examples):
        tokenized_inputs = tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)
        tokenized_inputs["labels"] = examples["label"]
        return tokenized_inputs

    tokenized_datasets = train_dataset.map(tokenize_and_align_labels, batched=True, remove_columns=train_dataset.column_names)
    return tokenized_datasets

# ฟังก์ชันสำหรับ fine-tune โมเดล
def fine_tune_model(train_dataset):
    training_args = TrainingArguments(
        output_dir="./results",
        num_train_epochs=3,
        per_device_train_batch_size=8,
        per_device_eval_batch_size=32,
        warmup_steps=100,
        weight_decay=0.01,
        logging_dir="./logs",
        logging_steps=10,
        evaluation_strategy="steps",
        eval_steps=500,
        save_steps=500,
        load_best_model_at_end=True,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=train_dataset.select(range(len(train_dataset) // 10)),  # Use 10% of data for evaluation
    )

    trainer.train()

    # Save the fine-tuned model
    trainer.save_model("./results")
    tokenizer.save_pretrained("./results")

# ฟังก์ชันสำหรับการวิเคราะห์ความรู้สึก
def analyze_sentiment(text):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)
    probs = outputs.logits.softmax(dim=-1)
    predicted_class = probs.argmax().item()
    label_map = {0: "negative", 1: "neutral", 2: "positive"}
    return {"label": label_map[predicted_class], "score": probs.max().item()}

# ฟังก์ชันหลัก
if __name__ == "__main__":
    print("เริ่มการเตรียมข้อมูล...")
    train_dataset = prepare_data()
    print("เตรียมข้อมูลเสร็จสิ้น")

    print("เริ่มการ fine-tune โมเดล...")
    fine_tune_model(train_dataset)
    print("Fine-tune โมเดลเสร็จสิ้น")

    # โหลดโมเดลและ tokenizer ที่ทำการ fine-tune แล้ว
    tokenizer = AutoTokenizer.from_pretrained("./results")
    model = AutoModelForSequenceClassification.from_pretrained("./results")

    # ทดสอบการวิเคราะห์ความรู้สึก
    test_texts = [
        "วันนี้อากาศดีมาก ฉันมีความสุขจัง",
        "เบื่อจังเลย ไม่อยากทำอะไรเลย",
        "อาหารร้านนี้รสชาติธรรมดามาก ไม่แย่แต่ก็ไม่อร่อย"
    ]

    print("\nทดสอบการวิเคราะห์ความรู้สึก:")
    for text in test_texts:
        result = analyze_sentiment(text)
        print(f"ข้อความ: {text}")
        print(f"ผลลัพธ์: {result}\n")
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, Trainer, TrainingArguments
from datasets import load_dataset, Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pythainlp.tokenize import word_tokenize as thai_word_tokenize
from pythainlp.corpus.common import thai_stopwords
from collections import Counter

# 1. Hugging Face Model Hub และ Transformers library
model_name = "airesearch/wangchanberta-base-att-spm-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3)

# 2. ฟังก์ชันสำหรับเตรียมข้อมูล
def prepare_data():
    dataset = load_dataset("sepidmnorozy/Thai_sentiment")
    train_dataset = dataset['train']

    print("Columns in the dataset:", train_dataset.column_names)
    print("Sample data:", train_dataset[0])

    def convert_label(example):
        label_map = {"positive": 1, "negative": 0}
        if isinstance(example["label"], str):
            example["label"] = label_map.get(example["label"].lower(), 1)
        return example

    train_dataset = train_dataset.map(convert_label)

    def tokenize_and_align_labels(examples):
        tokenized_inputs = tokenizer(examples["text"], padding="max_length", truncation=True, max_length=128)
        tokenized_inputs["labels"] = examples["label"]
        return tokenized_inputs

    tokenized_datasets = train_dataset.map(tokenize_and_align_labels, batched=True, remove_columns=train_dataset.column_names)
    return tokenized_datasets

# 3. ฟังก์ชันสำหรับ fine-tune โมเดล
def fine_tune_model(train_dataset):
    training_args = TrainingArguments(
        output_dir="./results",
        num_train_epochs=5,
        per_device_train_batch_size=64,
        per_device_eval_batch_size=64,
        warmup_steps=100,
        weight_decay=0.01,
        logging_dir="./logs",
        logging_steps=10,
        evaluation_strategy="steps",
        eval_steps=2500,
        save_steps=2500,
        load_best_model_at_end=True,
    )

    trainer = Trainer(
        model=model,
        args=training_args,
        train_dataset=train_dataset,
        eval_dataset=train_dataset.select(range(len(train_dataset) // 10)),  # ใช้ 10% ของข้อมูลสำหรับ evaluation
    )

    trainer.train()

    # บันทึกโมเดลที่ fine-tune แล้ว
    trainer.save_model("./results")
    tokenizer.save_pretrained("./results")

# 4. ฟังก์ชันวิเคราะห์ความรู้สึก
def analyze_sentiment(text):
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=128)
    inputs = {k: v.to(device) for k, v in inputs.items()}

    with torch.no_grad():
        outputs = model(**inputs)
    probs = outputs.logits.softmax(dim=-1)
    predicted_class = probs.argmax().item()
    label_map = {0: "NEGATIVE", 1: "POSITIVE"}
    return {"label": label_map[predicted_class], "score": probs.max().item()}

# 5. ฟังก์ชันสำหรับการวิเคราะห์คำสำคัญ
def extract_keywords(text, top_n=10):
    words = thai_word_tokenize(text)
    words = [word.lower() for word in words if word.lower() not in thai_stopwords()]
    word_freq = Counter(words)
    return word_freq.most_common(top_n)

# 6. ฟังก์ชันสร้างกราฟแสดงผลการวิเคราะห์
def plot_sentiment_distribution(sentiments):
    plt.figure(figsize=(10, 6))
    sns.countplot(x=sentiments)
    plt.title("การกระจายของความรู้สึกในรีวิว")
    plt.xlabel("ความรู้สึก")
    plt.ylabel("จำนวน")
    plt.savefig("sentiment_distribution.png")
    plt.close()

# 7. ฟังก์ชันสรุปผลการวิเคราะห์
def summarize_analysis(df):
    total_reviews = len(df)
    positive_reviews = len(df[df['sentiment'] == 'POSITIVE'])
    negative_reviews = len(df[df['sentiment'] == 'NEGATIVE'])

    print(f"จำนวนรีวิวทั้งหมด: {total_reviews}")
    print(f"รีวิวเชิงบวก: {positive_reviews} ({positive_reviews/total_reviews*100:.2f}%)")
    print(f"รีวิวเชิงลบ: {negative_reviews} ({negative_reviews/total_reviews*100:.2f}%)")

    all_text = " ".join(df['text'])
    top_keywords = extract_keywords(all_text)
    print("\nคำสำคัญที่พบบ่อย:")
    for word, count in top_keywords:
        print(f"{word}: {count}")

# 8. ฟังก์ชันหลักของโปรแกรม
def main():
    print("เริ่มการเตรียมข้อมูล...")
    train_dataset = prepare_data()
    print("เตรียมข้อมูลเสร็จสิ้น")

    print("เริ่มการ fine-tune โมเดล...")
    fine_tune_model(train_dataset)
    print("Fine-tune โมเดลเสร็จสิ้น")

    # โหลดโมเดลและ tokenizer ที่ทำการ fine-tune แล้ว
    tokenizer = AutoTokenizer.from_pretrained("./results")
    model = AutoModelForSequenceClassification.from_pretrained("./results")

    # โหลดข้อมูลเพื่อวิเคราะห์
    dataset = load_dataset("sepidmnorozy/Thai_sentiment")
    df = pd.DataFrame(dataset['train'])

    # วิเคราะห์รีวิว
    df['sentiment'] = df['text'].apply(lambda x: analyze_sentiment(x)['label'])

    # สร้างกราฟ
    plot_sentiment_distribution(df['sentiment'])

    # สรุปผลการวิเคราะห์
    summarize_analysis(df)

if __name__ == "__main__":
    main()
```
