```python
from qdrant_client import models, QdrantClient
from sentence_transformers import SentenceTransformer
```

```python
encoder = SentenceTransformer('all-MiniLM-L6-v2') # Model to create embeddings
```

```python
df = df.drop(columns='Unnamed: 0')
df
```

```python
# Convert DataFrame to a list of dictionaries
documents = df.to_dict(orient='records')
documents
```

```python
qdrant = QdrantClient(":memory:") # Create in-memory Qdrant instance
```

```python
# Create collection to store books
qdrant.recreate_collection(
    collection_name="test",
    vectors_config=models.VectorParams(
        size=encoder.get_sentence_embedding_dimension(), # Vector size is defined by used model
        distance=models.Distance.COSINE
    )
)
```

```python
documents
```

```python
# Let's vectorize descriptions and upload to qdrant

qdrant.upload_records(
    collection_name="test",
    records=[
        models.Record(
            id=idx,
            vector=encoder.encode(doc["text"]).tolist(),
            payload=doc
        ) for idx, doc in enumerate(documents)
    ]
)
```

```python
documents
```

```python
# Let's now search for something

hits = qdrant.search(
    collection_name="test",
    query_vector=encoder.encode("[AUTO][INET][INC-139385][INET_HealthCare][High (P2)] CrowdStrike - New sensor issue - machine DESKTOP-69UI76L กรุณาส่งรายละเอียดด้านล่างให้กับ Team [INET_CSIRT] และ cc [INET_CO-CSIRT] เพื่อตรวจสอบและแก้ไขปัญหาต่อไป").tolist(),
    limit=5
)
for hit in hits:
  print(hit.payload, "score:", hit.score)
```
