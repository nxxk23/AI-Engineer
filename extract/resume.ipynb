```python
import re
import deepcut

def custom_tokenize(text):
    tokens = re.split(r'(\s+|\n+)', text)
    # Remove empty tokens that may result from splitting
    tokens = [token for token in tokens if token.strip()]
    return tokens

def reconstruct_text_from_tokens(tokens):
    # Join the tokens back into a single string
    return ''.join(tokens)
```

```python
# Tokenize the text
tokens = custom_tokenize(text)
print(tokens)
```

```python
reconstructed_text = reconstruct_text_from_tokens(tokens)
print(reconstructed_text)
```

```python
tokens2 = deepcut.tokenize(reconstructed_text)
print(tokens2)
```

```python
def tokens_to_str(tokens):
    # Convert list of tokens to string format
    token_str = "[" + ", ".join(f"'{token}'" for token in tokens) + "]"
    return f'""{token_str}"""'

formatted_str = tokens_to_str(tokens2)
print(formatted_str)
```

```python
print(thaispellcheck.check(formatted_str))
```

```python
import fitz  # PyMuPDF
import pytesseract
from PIL import Image
import io

# Ensure Tesseract is set up correctly
pytesseract.pytesseract.tesseract_cmd = "/usr/bin/tesseract"

# Open the PDF file
pdf_path = "/content/drive/MyDrive/AIEngineer/resume/Resume ba2.pdf"
pdf_document = fitz.open(pdf_path)

# Extract Thai text from the first page using OCR
first_page = pdf_document.load_page(0)

# Convert the first page to an image
pix = first_page.get_pixmap(dpi=1100)  # Increase DPI for better accuracy
img = Image.open(io.BytesIO(pix.tobytes("png")))

# Use OCR to extract Thai text from the image
text = pytesseract.image_to_string(img, lang='tha+eng')

# Close the PDF document
pdf_document.close()

print("Extracted Thai Text:")
print(text)
```

```python
print(thaispellcheck.check(text))
```

```python
correct1 = thaispellcheck.check(text,autocorrect=True)
print(correct1)
```

```python
import cv2
import os

# Define the directory containing the images
image_directory = "/content/drive/MyDrive/AIEngineer/resume/images"

# Define the image file
image_file = "page_1.png"
image_path = os.path.join(image_directory, image_file)

# Load the image
image = cv2.imread(image_path)

if image is None:
    print(f"Failed to load image: {image_path}")
else:
    # Convert to grayscale without resizing
    gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # Save the processed image
    output_path = os.path.join(image_directory, f"processed_{image_file}")
    cv2.imwrite(output_path, gray_image)

    print(f"Processed image saved to {output_path}")
```

```python
import easyocr

# Create an EasyOCR Reader
reader = easyocr.Reader(['th', 'en'])

# Perform OCR on an image
result = reader.readtext('/content/drive/MyDrive/AIEngineer/resume/images/page_1.png')

# Print the results
for detection in result:
    print(detection)
```

```python
print(result)
```

```python
def extract_text_from_ocr(ocr_data):
    sorted_data = sorted(ocr_data, key=lambda x: x[0][0][1])
    plain_text = ""
    for _, text, _ in sorted_data:
        plain_text += text.strip() + "\n"  # Add newline to separate lines of text

    return plain_text.strip()

# Extract plain text
plain_text = extract_text_from_ocr(result)

# Print the extracted text
print(plain_text)
```
