```python
import re
import deepcut

def custom_tokenize(text):
    tokens = re.split(r'(\s+|\n+)', text)
    # Remove empty tokens that may result from splitting
    tokens = [token for token in tokens if token.strip()]
    return tokens

def reconstruct_text_from_tokens(tokens):
    # Join the tokens back into a single string
    return ''.join(tokens)
```

```python
# Tokenize the text
tokens = custom_tokenize(text)
print(tokens)
```

```python
reconstructed_text = reconstruct_text_from_tokens(tokens)
print(reconstructed_text)
```

```python
tokens2 = deepcut.tokenize(reconstructed_text)
print(tokens2)
```

```python
def tokens_to_str(tokens):
    # Convert list of tokens to string format
    token_str = "[" + ", ".join(f"'{token}'" for token in tokens) + "]"
    return f'""{token_str}"""'

formatted_str = tokens_to_str(tokens2)
print(formatted_str)
```

```python
print(thaispellcheck.check(formatted_str))
```
