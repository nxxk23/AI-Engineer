```python
import pandas as pd
data1 = pd.read_csv('/content/drive/MyDrive/AIEngineer/LATEST/RYH_DrNote_202412.csv',  encoding='TIS-620')
data2 = pd.read_csv('/content/drive/MyDrive/AIEngineer/LATEST/RYH_PatIncom_202412.csv',  encoding='TIS-620')
# data1.head()
```

```python
# Pivot ข้อมูลให้ IncomeCode แต่ละตัวเป็นคอลัมน์
pivot_data = data2.pivot(index="ID", columns="IncomeCode", values="Amt").reset_index()

# เปลี่ยนชื่อคอลัมน์ให้อ่านง่ายขึ้น
pivot_data.columns = ["ID"] + [f"Amt_{int(code)}" for code in pivot_data.columns[1:]]

# แทนค่าที่เป็น NaN (ไม่มีค่ารักษานั้น) ด้วย 0
pivot_data.fillna(0, inplace=True)
pivot_data.head()
```

```python
# รวมข้อมูลผู้ป่วย (Data1) กับค่าใช้จ่ายที่ Pivot แล้ว
merged_data = pivot_data.merge(data1, on="ID", how="left")
merged_data.head()
```
