```python
import pandas as pd
cut = pd.read_csv('/content/drive/MyDrive/AIEngineer/fraud/cut.csv')
cut
```

## **split diagnosis type**

```python
import ast
# Convert `icd_10` column from string to list of dictionaries
cut['icd_10'] = cut['icd_10'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Function to extract dx_types and remove the dots from codes
def extract_dx_types(row):
    result = {}
    for item in row:
        # Remove the dot in the ICD-10 code
        cleaned_code = item['code'].replace('.', '')
        result[item['dx_type']] = cleaned_code
    return result

# Apply the function to extract dx_types and create a DataFrame
extracted_data = cut['icd_10'].apply(extract_dx_types)
extracted_df = pd.DataFrame(extracted_data.tolist())
extracted_df
```

```python
cut = pd.concat([cut, extracted_df], axis=1)
cut
```

## **identify patients who have the exact same diagnosis combination**

```python
# Step 1: Convert `icd_10` column from string to list of dictionaries
cut['icd_10'] = cut['icd_10'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)
cut['billing'] = cut['billing'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Step 2: Calculate total billing
def calculate_total_billing(billing_list):
    return sum(float(item['net_amount']) for item in billing_list)

cut['total_billing'] = cut['billing'].apply(calculate_total_billing)

# Step 3: Clean ICD-10 codes by removing dots and extract diagnosis combinations
def get_cleaned_diagnosis_combination(row):
    codes = [item['code'].replace('.', '') for item in row]  # Remove dots from ICD-10 codes
    return tuple(sorted(codes))  # Sort and return as a tuple

cut['diagnosis_combination'] = cut['icd_10'].apply(get_cleaned_diagnosis_combination)

# Step 4: Group by diagnosis_combination and calculate aggregated stats, including the list of total_billing
grouped = (
    cut.groupby('diagnosis_combination')['total_billing']
    .agg(['mean', 'count', list])  # Add `list` to include all total_billing values for the group
    .reset_index()
)

# Rename the columns for better readability
grouped.rename(columns={'list': 'billing_list'}, inplace=True)
```
