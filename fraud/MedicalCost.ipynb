```python
# Step 1: Convert diagnosis column to Python objects
df['icd10'] = df['diagnosis'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Step 2: Function to process and split rows for multiple dx_types
def split_diagnoses(row):
    diagnoses = row['icd10']
    dx_dict = {}
    rows = []

    # Group diagnoses by dx_type
    for diag in diagnoses:
        dx_type = diag['dx_type']
        if dx_type not in dx_dict:
            dx_dict[dx_type] = []
        dx_dict[dx_type].append(diag['icd10'])

    # Create rows for each combination of dx_type
    from itertools import product
    keys, values = zip(*dx_dict.items())
    for combination in product(*values):  # Cartesian product of all dx_type combinations
        new_row = row.copy()
        for i, key in enumerate(keys):
            new_row[key] = combination[i]
        rows.append(new_row)
    return rows

# Step 3: Apply the splitting function
processed_rows = []
for _, row in df.iterrows():
    processed_rows.extend(split_diagnoses(row))

# Step 4: Create the final DataFrame
final_df = pd.DataFrame(processed_rows)
```

```python
import json
import re

def process_item(item_data):
    try:
        # Attempt default replacement
        return json.loads(item_data.replace("'", "\""))
    except json.JSONDecodeError:
        # Fallback: Replace problem single quote to avoid JSONDecodeError
        item_data = re.sub(r"([{\[])'", r"\1\"", item_data)  # Replace ' with " inside { and [
        item_data = re.sub(r"'\s*([}\]])", r'"\1', item_data) # Replace ' with " before } and ]
        return json.loads(item_data)
final_df['item_details'] = final_df['order_item'].apply(process_item)
item_expanded = final_df.explode('item_details')
item_expanded = item_expanded.reset_index(drop=True)
item_expanded = pd.concat([
    item_expanded.drop(['item_details'], axis=1),
    pd.json_normalize(item_expanded['item_details'])
], axis=1)

# Ensure numeric fields are properly cast
item_expanded['initial'] = pd.to_numeric(item_expanded['initial'], errors='coerce')
item_expanded['item_amount'] = pd.to_numeric(item_expanded['item_amount'], errors='coerce')

item_expanded.head()
```

```python
# df['icd10'] = df['diagnosis'].apply(lambda x: [d['icd10'] for d in x])

df['icd10'] = df['diagnosis'].apply(lambda x: next((d['icd10'] for d in x if d['dx_type'] == 'PP'), None))
df['dx_name'] = df['diagnosis'].apply(lambda x: next((d.get('dx_name') for d in x if d['dx_type'] == 'PP'), None))
```

```python
def safe_float(value):
    try:
        return float(value.strip()) if value.strip() else None
    except:
        return None

df['heart_rate'] = df['vital_sign'].apply(
    lambda x: safe_float(x[0]['heart_rate']) if isinstance(x, list) and len(x) > 0 and 'heart_rate' in x[0] else None
)
df['pain_score'] = df['vital_sign'].apply(
    lambda x: safe_float(x[0]['pain_score']) if isinstance(x, list) and len(x) > 0 and 'pain_score' in x[0] else None
)
df['systolic_bp'] = df['vital_sign'].apply(
    lambda x: safe_float(x[0]['systolic_bp']) if isinstance(x, list) and len(x) > 0 and 'systolic_bp' in x[0] else None
)
df['diastolic_bp'] = df['vital_sign'].apply(
    lambda x: safe_float(x[0]['diastolic_bp']) if isinstance(x, list) and len(x) > 0 and 'diastolic_bp' in x[0] else None
)
df['temperature'] = df['vital_sign'].apply(
    lambda x: safe_float(x[0]['temperature']) if isinstance(x, list) and len(x) > 0 and 'temperature' in x[0] else None
)
df['respiratory_rate'] = df['vital_sign'].apply(
    lambda x: safe_float(x[0]['respiratory_rate']) if isinstance(x, list) and len(x) > 0 and 'respiratory_rate' in x[0] else None
)
df['oxygen_saturation'] = df['vital_sign'].apply(
    lambda x: safe_float(x[0]['oxygen_saturation']) if isinstance(x, list) and len(x) > 0 and 'oxygen_saturation' in x[0] else None
)
df['vital_sign_entry_datetime'] = df['vital_sign'].apply(
    lambda x: x[0]['vital_sign_entry_datetime'] if isinstance(x, list) and len(x) > 0 and 'vital_sign_entry_datetime' in x[0] else None
)

df = df.drop(columns=['vital_sign'])
```

```python
def calculate_severity_index(row):
    severity = 0
    if row['heart_rate'] > 100:
        severity += 1
    if row['systolic_bp'] > 140:
        severity += 1
    if row['diastolic_bp'] > 90:
        severity += 1
    if row['pain_score'] > 5:
        severity += 1
    if row['temperature'] > 39:
        severity += 1
    if row['respiratory_rate'] > 20 or row['respiratory_rate'] < 12:
        severity += 1
    if row['oxygen_saturation'] < 90:
        severity += 1
    return severity
df['Severity Index'] = df.apply(calculate_severity_index, axis=1)
```

## **billing**

```python
# Function to expand icd10 and billing data
billing_df = []
for _, row in df.iterrows():
    icd10_list = row['icd10']
    billing_list = row['billing']

    # Repeat each billing entry for each icd10
    for icd10 in icd10_list:
        for billing in billing_list:
            billing_df.append({
                'illnes_type': row['illnes_type'],
                'icd10': row['icd10'],
                'simb_billing_code': billing.get('simb_billing_code'),
                'local_billing_code': billing.get('local_billing_code'),
                'local_billing_name': billing.get('local_billing_name'),
                'billing_initial': billing.get('billing_initial'),
            })

# Create DataFrame from the expanded billing data
billing_df = pd.DataFrame(billing_df)
```

```python

# Function to expand icd10 and order_items
order_item_data = []
for _, row in df.iterrows():
    icd10_list = row['icd10']
    order_items = row['order_item']

    # Repeat each order item for each icd10
    for icd10 in icd10_list:
        for item in order_items:
            order_item_data.append({
                'illnes_type': row['illnes_type'],
                'icd10': row['icd10'],
                'item_id': item['item_id'],
                'item_name': item['item_name'],
                'initial': item['initial']
            })

# Create DataFrame from the expanded data
order_item_df = pd.DataFrame(order_item_data)
```
